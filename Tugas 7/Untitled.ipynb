```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('titanic.csv')
dataset
```

```python
data = dataset[['Sex', 'Age', 'Pclass', 'Fare', 'Survived']]
data
```

```python
temp_data = data.isna()
temp_data
```

```python
naIsTrue = temp_data[temp_data["Age"] == True]
naIsFalse = temp_data[temp_data["Age"] == False]

pos_missing_test = np.array(naIsTrue.index)
pos_not_missing = np.array(naIsFalse.index)
```

```python
pos_not_missing
```

```python
train_data = data.drop(pos_missing_test)
train_data = train_data[['Sex', 'Pclass', 'Fare', 'Survived']]
train_data
```

```python
train_label = data.drop(pos_missing_test)
train_label = train_label['Age']
train_label
```

```python
test_data = data.drop(pos_not_missing)
test_data = test_data[['Sex', 'Pclass', 'Fare', 'Survived']]
test_data
```

```python
test_data= test_data.replace('male', 1)
test_data = test_data.replace('female', 0)
test_data
```

```python
train_data= train_data.replace('male', 1)
train_data = train_data.replace('female', 0)
train_data
```

```python
def min_max_scaling(data):
    
    data_norm = data.copy()
    
    for column in data_norm.columns:
        data_norm[column] = (data_norm[column] - data_norm[column].min()) / (data_norm[column].max() - data_norm[column].min())
        
    return data_norm
```

```python
norm_train_data = min_max_scaling(train_data)
norm_train_data
```

```python
norm_test_data = min_max_scaling(test_data)
norm_test_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn import utils
```

```python
lab_enc = preprocessing.LabelEncoder()
encoded = lab_enc.fit_transform(train_label)
encoded
```

```python
kNN=KNeighborsClassifier(n_neighbors=3, weights='distance')
kNN.fit(norm_train_data, encoded)
class_result = kNN.predict(norm_test_data)
class_result
```

```python
index = 0
for i in (pos_missing_test):
    data['Age'][i] = class_result[index]
    index+=1
```
