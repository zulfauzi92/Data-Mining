```python
def min_max_scaling(data):
    
    data_norm = data.copy()
    
    for column in data_norm.columns:
        data_norm[column] = (data_norm[column] - data_norm[column].min()) / (data_norm[column].max() - data_norm[column].min())
        
    return data_norm
```

```python
data_minMax = min_max_scaling(data)

data_minMax.head()
```

```python
def z_score(data):
    
    data_norm = data.copy()
    
    for column in data_norm.columns:
        data_norm[column] = (data_norm[column] - data_norm[column].mean()) / data_norm[column].std()
        
    return data_norm
```

```python
data_zScore = z_score(data)
data_zScore.head()
```

```python
def sigmoid(data):
    
    data_norm = data.copy()
    
    for column in data_norm.columns:
        x = (data_norm[column] - data_norm[column].mean()) / data_norm[column].std()
        data_norm[column] = (1 - np.exp(-x)) / (1 + np.exp(-x))
        
    return data_norm
```

```python
data_sigmoid = sigmoid(data)
data_sigmoid.head()
```
