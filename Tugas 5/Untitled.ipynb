```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('titanic.csv')
dataset
```

```python
test_dataset = pd.read_csv('titanic_test.csv')
test_dataset
```

```python
test_label = pd.read_csv('titanic_testlabel.csv')
test_label
```

```python
train_data = dataset[['Sex', 'Age', 'Pclass', 'Fare']]
```

```python
train_data
```

```python
train_data= train_data.replace('male', 1)
train_data = train_data.replace('female', 0)
```

```python
train_data
```

```python
mean = train_data['Age'].mean()

train_data = train_data.replace(np.nan, mean)
```

```python
train_data
```

```python
test_data = test_dataset[['Sex', 'Age', 'Pclass', 'Fare']]
test_data
```

```python
test_data= test_data.replace('male', 1)
test_data = test_data.replace('female', 0)
test_data
```

```python
mean = test_data['Age'].mean()

test_data = test_data.replace(np.nan, mean)
test_data
```

```python
train_label=dataset['Survived']
train_label
```

```python
test_label = test_label['Survived']
test_label
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import graphviz
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(train_data, train_label)
class_result=dtc.predict(test_data)
```

```python
class_result
```

```python
acc=dtc.score(test_data, test_label)
```

```python
acc
```

```python
err=round((1-acc)*100, 2)
print('\n\nError ratio = ', err, '%')
```
