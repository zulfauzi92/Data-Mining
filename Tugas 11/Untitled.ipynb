```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('transaction.csv')
dataset
```

```python
data = dataset[dataset['Country'] == 'Germany']
data
```

```python
data['Year'] = pd.DatetimeIndex(data['InvoiceDate']).year
data['Month'] = pd.DatetimeIndex(data['InvoiceDate']).month
```

```python
data
```

```python
data = data[data['Year'] == 2011]
data
```

```python
data = pd.DataFrame(data[['Month', 'Qty']])
data
```

```python
totalQty = data.groupby('Month')['Qty'].sum()
totalQty
```

```python
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x=totalQty.index
y=totalQty.values
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
linreg=LinearRegression()
x=np.array(x).reshape(-1,1)
linreg.fit(x, y)
```

```python
next_x=13
next_x=np.array(next_x).reshape(-1,1)
pred_x=linreg.predict(next_x)
```

```python
print('\nPrediksi x \n', pred_x.item())
```

```python
plt.scatter(next_x, pred_x, c='red')
pred_y=linreg.predict(x)
plt.plot(x, pred_y)
plt.plot(x, y)
plt.show()
```

```python
MSE=mean_squared_error(y,pred_y)
print('\nMSE = ', MSE)
```
